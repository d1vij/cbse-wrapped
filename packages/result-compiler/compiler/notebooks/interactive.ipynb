```python
import pandas as pd
from pathlib import Path
from compiler.clean_results import clean_school_result
from compiler.utils.parse import parse_int
```

```python
raw_str = (Path(".").parent / "sample_data" / "dav.json").read_text()
print(raw_str[:20])
```

```python
from compiler.models.RawResponseModels import RawSchoolResultJsonModel

raw = RawSchoolResultJsonModel.model_validate_json(raw_str)
```

```python
subject_ser = pd.Series()

for student in raw.success:
    data = student.data
    for idx in range(1,7):
        code = parse_int(getattr(data, f"SUB{idx}"))
        name = getattr(data, f"SNAME{idx}")
        if code:
            subject_ser[code] = name

print(subject_ser)
```
